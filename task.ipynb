```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:\datasets\imdb_top_1000.csv")
```

```python
chosen_cols = ['Genre', 'IMDB_Rating', 'Director']
df = df[chosen_cols]
```

```python
df = df.dropna(axis=0)
```

```python
genre_counts = df['Genre'].value_counts()
```

```python
most_popular_genre = genre_counts.index[0]
```

```python
genre_counts_df = pd.DataFrame(genre_counts).reset_index()
genre_counts_df.columns = ['Genre', 'Count']
```

```python
genre_counts_df.plot(x='Genre', y='Count', kind='bar')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.title('Most Popular Genre')
plt.show()
```

```python
genre_counts_df =genre_counts_df.head()
genre_counts_df.plot(x='Genre', y='Count', kind='bar')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.title('Most Popular Genre')
plt.show()
```

```python
director_counts = df['Director'].value_counts()
```
